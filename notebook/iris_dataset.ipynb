# testing our Neural network on the iris data set

To validate our neural network class we will utlize this notebook to train our neural network class on the most classic of datasets the Iris data set which consisits of 3 different types of Irises.
This data set is built into sklearn and more information can be found here [Iris_Dataset](https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html)

```python
# importing our data and our network
import os
import sys

import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

module_path= os.path.abspath(os.path.join('..',"python"))
if module_path not in sys.path:
    sys.path.append(module_path)
from neural_net import NeuralNetwork
```

loading data set

```python
iris = datasets.load_iris(as_frame=True)
x = iris.data
y = iris.target

x[:5]
```

```python
y[:5]
```

we will dummy our y's for training

```python
y_dummy = pd.get_dummies(y)
```

creating our test train splits

```python
X_train, X_test, y_train, y_test,Y_nd_train,Y_nd_test  = train_test_split(x.values,y_dummy.values,y ,train_size=.2,random_state=100)
y_train
```

initializing our model and returning an initial score

```python
model = NeuralNetwork(4,3,3,) # we will kep the default sigmoid

#sainty checking our initial weight matrix
model.w
```

```python
# create  classification report func
def c_func(X,y):
    pred = [np.argmax(x) for x in model.predict(X)]
    return classification_report(y,pred, zero_division = 0)

print(c_func(X_train,Y_nd_train))
```

for our initial training we will run a single epoch `

```python
model.train(X_train,y_train)
```

```python
print(c_func(X_train,Y_nd_train))
```

we haven't really increased our performance yet we will run a few more epochs and see if that helps

```python
for x in range(5):
    #shuffle data
    X_shuffle, y_shuffle = shuffle(X_train,y_train)
    model.train(X_shuffle,y_shuffle)
```

```python
print(c_func(X_train,Y_nd_train))
```

our network is margninaly better lets see after 30 more epochs

```python
for x in range(30):
    #shuffle data
    X_shuffle, y_shuffle = shuffle(X_train,y_train)
    model.train(X_shuffle,y_shuffle)
```

```python
print(c_func(X_train,Y_nd_train))
```

now we are making progress lets add 100 epochs

```python
for x in range(100):
    #shuffle data
    X_shuffle, y_shuffle = shuffle(X_train,y_train)
    model.train(X_shuffle,y_shuffle)
```

```python
print(c_func(X_train,Y_nd_train))
```

97% is good but lets see how it does agains out testing set

```python
print(c_func(X_test,Y_nd_test))
```

still pretty good lets see how 1,000 more epochs effects this

```python
for x in range(1000):
    #shuffle data
    X_shuffle, y_shuffle = shuffle(X_train,y_train)
    model.train(X_shuffle,y_shuffle)
```

```python
print(c_func(X_train,Y_nd_train))
```

```python
print(c_func(X_test,Y_nd_test))
```

we have perfectly fit our training data and still improved our testing set
